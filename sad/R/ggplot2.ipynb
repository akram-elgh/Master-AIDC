```r
# scatterplot (good for two quantitative data )
ggplot(crickets, aes(x = temp,
                     y = rate,
                     color = species)
) +
  geom_point() +
  labs(
    x = "Temperature",
    y = "Rate",
    color = "Species",
    title = "Cricket chirps",
    caption = "Source : McDonald (2009)"
  )
```

```r
# modifying basic properties of the plot
ggplot(crickets, aes(x = temp,
                     y = rate,
                    )
) +
  geom_point(
    color = "red",
    size = 3,
    alpha = 0.4, # opacity
    shape = "square", # shape of dots
  ) +
  labs(
    x = "Temperature",
    y = "Rate",
    color = "Species",
    title = "Cricket chirps",
    caption = "Source : McDonald (2009)"
  )
```

```r
# adding linear regression layer

ggplot(crickets, aes(
  x = temp,
  y = rate
)) +
  geom_point() +
  geom_smooth() +
  labs(
    x = "Temperature",
    y = "Rate",
    color = "Species",
    title = "Cricket chirps",
    caption = "Source : McDonald (2009)"
  )
```

```r
# with linear method and without standard error (gray)
ggplot(crickets, aes(
  x = temp,
  y = rate
)) +
  geom_point() +
  geom_smooth(
    method = "lm",
    se = FALSE
  ) +
  labs(
    x = "Temperature",
    y = "Rate",
    color = "Species",
    title = "Cricket chirps",
    caption = "Source : McDonald (2009)"
  )
```

```r
# with linear regression for each species
# with linear method and without standard error (gray)
ggplot(crickets, aes(
  x = temp,
  y = rate,
  color = species
)) +
  geom_point() +
  geom_smooth(
    method = "lm",
    se = FALSE
  ) +
  labs(
    x = "Temperature",
    y = "Rate",
    color = "Species",
    title = "Cricket chirps",
    caption = "Source : McDonald (2009)"
  )
```

```r
# other plots
ggplot(crickets, aes(x = rate)) +
  geom_histogram(bins = 15)
```

```r
# other plots (good for quantitative data)
ggplot(crickets, aes(x = rate)) +
  geom_freqpoly(bins = 15)
```

```r
# bar charts (good for qualitative data)
ggplot(crickets, aes(
  x = species,
  fill = species
)) +
  geom_bar(show.legend = FALSE)
```

```r
# boxplot (good for combining qualitative and quantitative data)
# other plots
ggplot(crickets, aes(x = species,
                     y = rate,
                     color = species)) +
  geom_boxplot(show.legend = FALSE)
```

```r
# facet_wrap
ggplot(crickets, aes(x = rate, fill = species)) +
  geom_histogram(show.legend = FALSE, bins = 15) +
  facet_wrap(~species) +
  theme_minimal()
```

```r
library(tidytuesdayR)
tuesdata <- tidytuesdayR::tt_load(2021, week = 41)
nurses <- tuesdata$nurses
```

```r
head(nurses)
```

```r
nurses_2020 <- nurses %>%
  filter(Year == 2020)

nurses_2020 <- slice_max(nurses_2020,
  order_by = nurses_2020$"Total Employed RN", n = 20
)
```

```r
ggplot(nurses_2020, aes(
  x = nurses_2020$"Total Employed RN",
  y = fct_reorder(State, nurses_2020$"Total Employed RN"),
  fill = State
)) +
  geom_col(show.legend = FALSE) +
  geom_text(aes(label = nurses_2020$"Total Employed RN")) +
  theme_minimal()
```
