# Data Summary

## Exploration

```r
protein_data <- read.csv("data/protein.csv")

# Show the first 5 rows
head(protein_data)
```

```r
# Number of rows and columns in the data
dimensions <- dim(protein_data)
n_rows <- dimensions[1]
n_columns <- dimensions[2]

cat("Number of rows:", n_rows, "\n")
cat("Number of columns:", n_columns, "\n")
```

```r
# Let's have a quick overview of the data
str(protein_data)
```

```r
# Check for missing values
colSums(is.na(protein_data))
```

```r
# Install & Load the corrr package
library('corrr')
```

```r
# Select numerical columns
numerical_data <- protein_data[,2:10]
head(numerical_data)
```

```r
# Normalize the data
data_normalized <- scale(numerical_data)
head(data_normalized)
```

```r
# Compute the correlation matrix
corr_matrix <- cor(data_normalized)
corr_matrix
```

```r
# visualisation using corrplot
library(corrplot)
corrplot(corr_matrix, method = "color")
```

```r
# Install and load the package for plotting the matrix as heatmap.
library(ggcorrplot)

ggcorrplot(corr_matrix)
```

## Apply PCA

```r
library("FactoMineR")
library("factoextra")
```

```r
# Apply PCA
data.pca <- princomp(corr_matrix)

# Summary of the PCA analysis
summary(data.pca)
```

```r
str(data.pca)
```

```r
# Show the result
summary(data.pca)
```

```r
# Check the loadings of the first two components.
data.pca$loadings[, 1:2]
```
