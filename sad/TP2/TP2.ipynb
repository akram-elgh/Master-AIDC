```r
xi <- 1:12
yi <- c(40, 42, 44, 45, 48, 50, 52, 55, 58, 63, 68, 70)
```

```r
plot(xi, yi)
```

```r
cor.test(xi, yi)
```

```r
qqnorm(yi)
qqline(yi, col = "darkgreen")
qqplot(yi)
```

```r
regxy <- lm(yi ~ xi)
regxy
```

```r
plot(xi, yi)
abline(regxy, col = "red")
```

```r
summary(regxy)
```

## tester l'independence des résidus    

```r
acf(residuals(regxy), main = "reg xy")
library(car)
durbinWatsonTest(regxy)
# dwtest(regxy)
```

## test de normalité  

```r
plot(regxy,2)
shapiro.test(residuals(regxy))
residualPlot((regxy))
residualPlots((regxy))
```

## Tester l'homogenite des residus

```r
plot(regxy, 3)
ncvTest(regxy)
```

## Intervall de confiance

```r
plot(regxy,1)
confint(regxy)
```
